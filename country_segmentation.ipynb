# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
scaled_df =scaler.fit_transform(df)
```

```python
# finding optimal clusters
from sklearn.cluster import KMeans
wcss =[]
for i in range(1,10):
  kmeans =KMeans(n_clusters =i, random_state =42)
  kmeans.fit(scaled_df)
  wcss.append(kmeans.inertia_)

plt.plot(range(1,10), wcss, marker ='o')
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
#Applying clustering
kmeans =KMeans(n_clusters =3, random_state =42)
cluster =kmeans.fit_predict(scaled_df)
df['clusters'] =cluster
```

```python
df.groupby('clusters').mean()
```

```python
df['clusters'].value_counts()
```

**PCA**

```python
from sklearn.decomposition import PCA
pca =PCA(n_components =2)
pca_data =pca.fit_transform(scaled_df)

df['PCA1'] =pca_data[:,0]
df['PCA2'] =pca_data[:,1]
```

```python
# plotting clusters
sns.scatterplot(x ='PCA1', y ='PCA2', data =df)
plt.title("country cluster (pca)")
plt.show()
```

```python
# validate clustering
from sklearn.metrics import silhouette_score
score =silhouette_score(scaled_df, cluster)
print("silhouette score:", score)
```

```python
for k in range(2,6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(scaled_df)
    score = silhouette_score(scaled_df, labels)
    print(f"K={k}, Score={score}")
```

Although K=5 gave a slightly better silhouette score, I selected K=3 because it provides more interpretable and meaningful segmentation aligned with real-world country development categories.

```python
kmeans_1 = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans_1.fit_predict(scaled_df)
```

```python
df.groupby('Cluster').mean(numeric_only =True)
```

```python
df['Category'] = df['Cluster'].map({
    0: 'Developing',
    1: 'Developed',
    2: 'Underdeveloped'
})
```

**Hierarchical** **clustering**

```python
# importing
from scipy.cluster.hierarchy import dendrogram, linkage

```

```python
# creating dendrogram
plt.figure(figsize =(12,6))
linked =linkage(scaled_df,method ='ward')
dendrogram(linked, truncate_mode ='lastp', p =20)
plt.title("Dendrogram")
plt.xlabel("countries")
plt.ylabel("distance")
plt.show()
```

```python
# using agglomerative clustering
from sklearn.cluster import AgglomerativeClustering
hc =AgglomerativeClustering(n_clusters =3)
hc_labels =hc.fit_predict(scaled_df)
df['HC_cluster'] =hc_labels
```

```python
df.groupby('HC_cluster').mean(numeric_only =True)
```

```python
df['HC_Category'] = df['HC_cluster'].map({
    0: 'Developed',
    1: 'Developing',
    2: 'Underdeveloped'
})
```

Cluster 0 represents developed countries characterized by high income and low child mortality, while Cluster 2 represents underdeveloped countries with low income and high mortality.

```python
pd.crosstab(df['Category'], df['HC_Category'])
```

Although cluster labels differed numerically between K-Means and Hierarchical clustering, both methods produced consistent groupings in terms of socio-economic characteristics.
